```python
##get everything into dataframes for cross referencing
#PRJ files give the application number (needed for project abstracts)
import pandas as pd
def readMyProj(filename):
    #imported as object to save memory, will convert to appropriate data types as needed
    df = pd.read_csv(filename, encoding='cp1252', dtype = object) 
    # note not utf8 like other files
    #details for analysis : total cost -how much does a publication cost? STUDY_SECTION_NAME who is giving out the most $
    #for fun Org_city and Org_name- where does the money go? Application type (#1 is new) for time to pub
    keepColumns = ['APPLICATION_ID', 'ADMINISTERING_IC', 'APPLICATION_TYPE', 'CORE_PROJECT_NUM', 'FULL_PROJECT_NUM', 'ORG_NAME',
                'ORG_STATE', 'STUDY_SECTION_NAME', 'DIRECT_COST_AMT','INDIRECT_COST_AMT','TOTAL_COST']
    #new truncated DataFrame
    dfProjDetails = df[keepColumns]
    return dfProjDetails

dfProjDetails = readMyProj('RePORTER_PRJ_C_FY2015.csv')
```

## I am going to similarly manipulate the publication data now so that the maps can be visualized together

```python
import pandas as pd
def pubfreq(filename):
    #this is the file which links 2016 publications to grant numbers
    link = pd.read_csv(filename)
    pubsPerGrant = pd.DataFrame(link['PROJECT_NUMBER'].copy())
    #assigns each link a frequency value
    pubsPerGrant['FREQ']= 1
    newppG = pd.DataFrame(pubsPerGrant.groupby(by=['PROJECT_NUMBER'])['FREQ'].sum())
    return newppG
newppG = pubfreq("RePORTER_PUBLNK_C_2015.csv")
```

```python
##the number of publications per grant was joined to the larger file with the grant information
def pubcount(projectFile, pubsPerGrant):
    projwithPubs = pd.DataFrame(projectFile.copy())
    projwithPubs.set_index(['CORE_PROJECT_NUM'], inplace = True)
    howManyPubs = projwithPubs.join(pubsPerGrant)
    return howManyPubs
howManyPubs = pubcount(dfProjDetails, newppG)
```

```python
def statePubs(pubFile):
    pubFile['FREQ'] = pd.to_numeric(pubFile['FREQ'], errors = "coerce")
    pivotPubS = pd.DataFrame(pubFile.pivot_table(values = 'FREQ', index = 'ADMINISTERING_IC', 
                                                    columns= 'ORG_STATE', aggfunc = (np.sum)))
    pivotPubS.to_csv('pubsStateSection.csv',header ='column_names')
    stateSum = pivotPubS.sum(axis = 0)
    stateTotal= pd.concat([stateSum, s], axis = 1)
    pivotReady = pd.DataFrame(stateTotal.dropna())
    return(pivotReady)

pivotReady = statePubs(howManyPubs)
pivotReady.info()


```

```python
import vincent
vincent.core.initialize_notebook()
state_topo = "https://raw.githubusercontent.com/wrobstory/vincent_map_data/master/us_states.topo.json"

geo_data = [{'name': 'states',
             'url': state_topo,
             'feature': 'us_states.geo'}]
def visState(mapresult):
    #Standardize State names to match TopoJSON for keying
    mapresult = mapresult.dropna()
    mapresult.columns = ['Total/100000', 'NAME']
    names = [mapresult['NAME']]
    #define vis object
    vis = vincent.Map(data=mapresult, geo_data=geo_data, scale=800,
                      projection='albersUsa', data_bind='Total/100000', data_key='NAME',
                      map_key={'states': 'properties.NAME'}, brew='YlGnBu')
    #Custom threshold scale
    vis.scales[0].type='threshold'
    vis.scales[0].domain = [ 0, 300, 900, 2700, 8100, 24300 ]
    vis.legend(title=' x $100,000')
    vis.to_json('vega.json')
    return(vis.display())
visState(pivotMoney)
```
